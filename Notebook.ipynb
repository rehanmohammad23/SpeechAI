```python
def assign_label(pronun, fluency):
    avg_score = (pronun + fluency) / 2
    
    if avg_score < 25:
        return "Traditional"
    elif avg_score < 45:
        return "Average"
    elif avg_score < 65:
        return "Group Control"
    else:
        return "AI-Powered"
```

```python
df['target'] = df.apply(lambda x: assign_label(x['Pronun'], x['Fluency']), axis=1)
```

```python
import pandas as pd

df = pd.read_csv("final_dataset.csv")
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
df = df[['text', 'target']].dropna()
```

```python
from sklearn.utils import resample

df_balanced = pd.DataFrame()

max_samples = df['target'].value_counts().max()

for label in df['target'].unique():
    df_label = df[df['target'] == label]
    df_label_upsampled = resample(df_label,
                                 replace=True,
                                 n_samples=max_samples,
                                 random_state=42)
    df_balanced = pd.concat([df_balanced, df_label_upsampled])

print(df_balanced['target'].value_counts())
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

df_balanced['clean_text'] = df_balanced['text'].apply(clean_text)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_balanced['label'] = le.fit_transform(df_balanced['target'])
```

```python
import joblib
joblib.dump(le, 'models/label_encoder.pkl')
```

```python
from sklearn.model_selection import train_test_split

train_texts, test_texts, train_labels, test_labels = train_test_split(
    df_balanced['clean_text'],
    df_balanced['label'],
    test_size=0.3,
    stratify=df_balanced['label'],
    random_state=42
)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
```

# BERT

```python
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import torch

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

train_encodings = tokenizer(list(train_texts), truncation=True, padding=True, max_length=128)
test_encodings = tokenizer(list(test_texts), truncation=True, padding=True, max_length=128)

class Dataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels.iloc[idx], dtype=torch.long)
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = Dataset(train_encodings, train_labels)
test_dataset = Dataset(test_encodings, test_labels)

model_bert = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=4)

training_args = TrainingArguments(
    output_dir='./bert_results',
    num_train_epochs=5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    fp16=True,
    logging_dir='./logs'
)

trainer = Trainer(
    model=model_bert,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

trainer.train()
```

```python
bert_preds = trainer.predict(test_dataset).predictions.argmax(axis=1)

bert_acc = accuracy_score(test_labels, bert_preds)
bert_prec = precision_score(test_labels, bert_preds, average='weighted')
bert_rec = recall_score(test_labels, bert_preds, average='weighted')
bert_f1 = f1_score(test_labels, bert_preds, average='weighted')
```

# RoBERTa

```python
from transformers import RobertaTokenizer, RobertaForSequenceClassification

tokenizer_r = RobertaTokenizer.from_pretrained('roberta-base')

train_enc_r = tokenizer_r(list(train_texts), truncation=True, padding=True)
test_enc_r = tokenizer_r(list(test_texts), truncation=True, padding=True)

train_dataset_r = Dataset(train_enc_r, train_labels)
test_dataset_r = Dataset(test_enc_r, test_labels)

model_r = RobertaForSequenceClassification.from_pretrained("roberta-base", num_labels=4)

trainer_r = Trainer(
    model=model_r,
    args=training_args,
    train_dataset=train_dataset_r,
    eval_dataset=test_dataset_r
)

trainer_r.train()
```

```python
roberta_preds = trainer_r.predict(test_dataset_r).predictions.argmax(axis=1)

rob_acc = accuracy_score(test_labels, roberta_preds)
rob_prec = precision_score(test_labels, roberta_preds, average='weighted')
rob_rec = recall_score(test_labels, roberta_preds, average='weighted')
rob_f1 = f1_score(test_labels, roberta_preds, average='weighted')
```

# BERT Embeddings + ML

```python
from transformers import BertModel
import numpy as np

bert_model = BertModel.from_pretrained('bert-base-uncased')

def get_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    outputs = bert_model(**inputs)
    return outputs.last_hidden_state[:,0,:].detach().numpy()[0]

X = np.array([get_embedding(t) for t in df_balanced['clean_text']])
y = df_balanced['label'].values
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
```

# Hybrid Model

```python
from sklearn.svm import SVC

kernel = 'rbf'
gamma = 79.783
C = 3.7442
coef0 = 0.9021
degree = 3
svm = SVC(kernel=kernel, gamma=gamma, C=C, degree=degree, coef0=coef0)
```

```python
from sklearn.ensemble import RandomForestClassifier

parametros_rf = {
    'n_estimators': 1000,
    'criterion': 'entropy',
    'max_depth': None,
    'min_samples_split': 2,
    'min_samples_leaf': 1
}

rf = RandomForestClassifier(**parametros_rf)
```

```python
from sklearn.tree import DecisionTreeClassifier

max_depth_param = None
min_samples_split_param = 3
min_samples_leaf_param = 1
criterion_param = 'entropy'

dt = DecisionTreeClassifier(
    max_depth=max_depth_param,
    min_samples_split=min_samples_split_param,
    min_samples_leaf=min_samples_leaf_param,
    criterion=criterion_param
)
```

```python
from sklearn.ensemble import StackingClassifier, GradientBoostingClassifier

base_learners = [
    ('svm', svm),
    ('rf', rf),
    ('dt', dt)
]

meta_learner = GradientBoostingClassifier(
    n_estimators=100,     # learning_cycles
    learning_rate=1,      # learning_rate
    min_samples_leaf=7,
    random_state=42
)

stack = StackingClassifier(
    estimators=base_learners,
    final_estimator=meta_learner
)
```

```python
stack.fit(X_train, y_train)

y_pred_stack = stack.predict(X_test)
print("Hybrid Model\n", classification_report(y_test, y_pred_stack))
```

```python
hyb_acc = accuracy_score(y_test, y_pred_stack)
hyb_prec = precision_score(y_test, y_pred_stack, average='weighted')
hyb_rec = recall_score(y_test, y_pred_stack, average='weighted')
hyb_f1 = f1_score(y_test, y_pred_stack, average='weighted')
```

# Extract Audio Features

```python
import librosa
import numpy as np
import os

audio_folder = "speech_data/audio onepte-repeat sentence"

def extract_features(file_name):
    try:
        path = os.path.join(audio_folder, file_name)
        y, sr = librosa.load(path, sr=22050)
        
        mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)
        chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr).T, axis=0)
        contrast = np.mean(librosa.feature.spectral_contrast(y=y, sr=sr).T, axis=0)
        
        return np.hstack([mfcc, chroma, contrast])
    except:
        return np.zeros(59)  # fallback

X_audio = np.array([extract_features(f) for f in df_balanced['audio']])

print("Feature shape:", X_audio.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(df_balanced['target'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_audio = scaler.fit_transform(X_audio)
```

```python
# Save to a file
joblib.dump(scaler, 'models/scaler_audio.pkl')
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_audio, y, test_size=0.3, stratify=y, random_state=42
)
```

```python
def get_metrics(y_true, y_pred):
    return (
        accuracy_score(y_true, y_pred),
        precision_score(y_true, y_pred, average='weighted'),
        recall_score(y_true, y_pred, average='weighted'),
        f1_score(y_true, y_pred, average='weighted')
    )
```

# DL Model

```python
import torch
import torch.nn as nn
import torch.optim as optim

class AudioMLP(nn.Module):
    def __init__(self, input_dim, num_classes):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )
    
    def forward(self, x):
        return self.model(x)

model = AudioMLP(X_train.shape[1], len(le.classes_))

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

X_train_t = torch.tensor(X_train, dtype=torch.float32)
y_train_t = torch.tensor(y_train, dtype=torch.long)

# Training
for epoch in range(15):
    optimizer.zero_grad()
    outputs = model(X_train_t)
    loss = criterion(outputs, y_train_t)
    loss.backward()
    optimizer.step()
    
    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```

```python
y_test_t = torch.tensor(y_test, dtype=torch.long)
```

```python
X_test_t = torch.tensor(X_test, dtype=torch.float32)

outputs = model(X_test_t)
_, y_pred_dl = torch.max(outputs, 1)

y_pred_dl = y_pred_dl.numpy()
```

```python
dl_metrics = get_metrics(y_test_t, y_pred_dl)
```
